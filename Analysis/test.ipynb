```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_type_dict = \
    {"Date": str,
     "Department": str,
     "Course": str,
     "Language": str,
     "Tutor": str}
```

```python
data_frame = pd.read_csv("AggregateData/aggregate_data.tsv",
                         delimiter="\t",
                         dtype=data_type_dict)
```

```python
data_frame['JustDate'] = data_frame.apply(lambda row: row[0].split()[0], axis=1)
```

```python
print(data_frame.head())
```

```python
gb = data_frame.groupby('JustDate').size()
```

```python
dates_as_str = gb.index.tolist()
counts = gb.values
```

```python
plt.plot_date(x=dates_as_str, y=counts, fmt='r-')
plt.xticks(rotation=10)
```

```python
plt.title("Helping Instances per day: Fall 2016 Semester", fontsize=16)
plt.xlabel("Date", fontsize=16)
plt.ylabel("helping instance count", fontsize=16)
plt.show()
```
